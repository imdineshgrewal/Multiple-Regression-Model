```python
#dividing data into features and label
Y = df_work['cnt']
X = df_work.drop(['cnt'], axis=1)
```

```python
#Divide data into test and train
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .20, random_state = 42)
```

```python
#SVM
from sklearn.svm import SVR
np.random.seed(0)
clf = SVR(gamma='scale', C=1.0, epsilon=0.2)
clf.fit(X_train, Y_train)

predicted = clf.predict(X_test)
residual = Y_test - predicted

fig = plt.figure(figsize=(8,8))
ax1 = plt.subplot(211)
sns.distplot(residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=7)
plt.title('SVM Residual counts',fontsize=8)
plt.xlabel('Residual',fontsize=8)
plt.ylabel('Count',fontsize=8)

ax2 = plt.subplot(212)
plt.scatter(predicted, residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=7)
plt.xlabel('Predicted',fontsize=8)
plt.ylabel('Residual',fontsize=8)
plt.axhline(y=0)
plt.title('SVM Residual vs. Predicted',fontsize=8)

plt.show()

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(Y_test, predicted))
print('RMSE:')
print(rmse)

from sklearn.metrics import r2_score
print('rSquare: %.2f' % r2_score(Y_test, predicted))
```

```python
#KNN - k-Nearest Neighbors

from sklearn import neighbors
# the value of n_neighbors will be changed when we plot the histogram showing the lowest RMSE value
knn = neighbors.KNeighborsRegressor(n_neighbors=6)
knn.fit(X_train, Y_train)

predicted = knn.predict(X_test)
residual = Y_test - predicted

fig = plt.figure(figsize=(8,8))
ax1 = plt.subplot(211)
sns.distplot(residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=7)
plt.title('KNN Residual counts',fontsize=8)
plt.xlabel('Residual',fontsize=8)
plt.ylabel('Count',fontsize=8)

ax2 = plt.subplot(212)
plt.scatter(predicted, residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=7)
plt.xlabel('Predicted',fontsize=8)
plt.ylabel('Residual',fontsize=8)
plt.axhline(y=0)
plt.title('KNN Residual vs. Predicted',fontsize=8)

plt.show()

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(Y_test, predicted))
print('RMSE:')
print(rmse)
print('rSquare: %.2f' % r2_score(Y_test, predicted))
```

```python
#Decision tree

from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor(max_features='auto')
dtr.fit(X_train, Y_train)
predicted = dtr.predict(X_test)
residual = Y_test - predicted

fig = plt.figure(figsize=(8,8))
ax1 = plt.subplot(211)
sns.distplot(residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=7)
plt.title('DecisionT Residual counts',fontsize=8)
plt.xlabel('Residual',fontsize=8)
plt.ylabel('Count',fontsize=8)

ax2 = plt.subplot(212)
plt.scatter(predicted, residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=7)
plt.xlabel('Predicted',fontsize=8)
plt.ylabel('Residual',fontsize=8)
plt.axhline(y=0)
plt.title('DecisionT Residual vs. Predicted',fontsize=8)

plt.show()

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(Y_test, predicted))
print('RMSE:')
print(rmse)

print('rSquare: %.2f' % r2_score(Y_test, predicted))
```

```python
#Linear regression 

from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(X_train, Y_train)

predicted = regr.predict(X_test)
residual = Y_test - predicted

fig = plt.figure(figsize=(8,8))
ax1 = plt.subplot(211)
sns.distplot(residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=7)
plt.title('LinearR Residual counts',fontsize=8)
plt.xlabel('Residual',fontsize=8)
plt.ylabel('Count',fontsize=8)

ax2 = plt.subplot(212)
plt.scatter(predicted, residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=7)
plt.xlabel('Predicted',fontsize=8)
plt.ylabel('Residual',fontsize=8)
plt.axhline(y=0)
plt.title('LinearR Residual vs. Predicted',fontsize=8)

plt.show()

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(Y_test, predicted))
print('RMSE:')
print(rmse)

print('rSquare: %.2f' % r2_score(Y_test, predicted))
```

```python
#Gradient Boosting

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score

gbr = GradientBoostingRegressor(loss ='ls', max_depth=6)
gbr.fit (X_train, Y_train)
predicted = gbr.predict(X_test)
residual = Y_test - predicted

fig = plt.figure(figsize=(8,8))
ax1 = plt.subplot(211)
sns.distplot(residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=7)
plt.title('GBR Residual counts',fontsize=8)
plt.xlabel('Residual',fontsize=8)
plt.ylabel('Count',fontsize=8)

ax2 = plt.subplot(212)
plt.scatter(predicted, residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=7)
plt.xlabel('Predicted',fontsize=8)
plt.ylabel('Residual',fontsize=8)
plt.axhline(y=0)
plt.title('GBR Residual vs. Predicted',fontsize=8)

plt.show()

rmse = np.sqrt(mean_squared_error(Y_test, predicted))

######################### Cross Validation #########################
scores = cross_val_score(gbr, X, Y, cv=12) #12 

print('\nCross Validation Scores:')
print(scores)
print('\nMean Score:')
print(scores.mean())
print('\nRMSE:')
print(rmse)

print('rSquare: %.2f' % r2_score(Y_test, predicted))
```
